```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
df = np.array([[182,6,5,6,0],[1,148,10,20,2],[1,3,110,30,5],[2,10,26,130,0],[0,1,5,2,143]])
```

```python
df
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=df, display_labels=["tree", "areca", "grass", "bamboo", "not_veg"])
```

```python
disp.plot(include_values=True, cmap="viridis", ax=None, values_format="d")
plt.show()
```
